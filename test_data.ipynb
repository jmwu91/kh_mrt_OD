#### 高雄捷運每日各站分時資料
- 測試資料
- 資料時間: 2024.05.31

```python
import pandas as pd
```

```python
file_path_O = "./data/00354_sample.csv"
file_path_D = "./data/00355_sample.csv"

data_O = pd.read_csv(file_path_O)
data_D = pd.read_csv(file_path_D)
```

```python
# 調整資料
# 刪除第一個 row 並將第二個 row 當成 header
data_O.columns = data_O.iloc[0]  # 將第0個 row 當成 header
df_O = data_O[1:]  # 刪除第一個row（原來的 header 和第一個 row）

# 重設index
df_O.reset_index(drop = True, inplace = True)
df_O.head(10)
```

```python
data_D.columns = data_D.iloc[0]  # 將第0個 row 當成 header
df_D = data_D[1:]  # 刪除第一個row（原來的 header 和第一個 row）

# 重設index
df_D.reset_index(drop = True, inplace = True)
df_D.head(10)
```

```python
# 清理資料欄位
df_O.drop(["OriginStationName",
          "DestinationStationName", "InfoDate"], axis = 1, inplace = True)

df_D.drop(["OriginStationName",
          "DestinationStationName", "InfoDate"], axis = 1, inplace = True)
```

```python
# 調整時間欄位
# df_O
df_O.loc[:, 'TripHour'] = df_O['TripHour'].astype(int)
df_O.loc[:, 'adjusted_time'] = df_O['TripHour'].apply(lambda x: x if x >= 5 else x + 24)
df_O = df_O.sort_values(by = 'adjusted_time')

# df_D
df_D.loc[:, 'TripHour'] = df_D['TripHour'].astype(int)
df_D.loc[:, 'adjusted_time'] = df_D['TripHour'].apply(lambda x: x if x >= 5 else x + 24)
df_D = df_D.sort_values(by = 'adjusted_time')

# 清理空值
df_O = df_O.dropna(subset = ["HolderType"])
print(df_O.shape)
df_D = df_D.dropna(subset = ["HolderType"])
print(df_D.shape)
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df_D):
    # Change column type to object for column: 'Volume'
    df_D = df_D.astype({'Volume': 'int'})
    return df_D

df_D_clean = clean_data(df_D.copy())
df_D_clean.head()

def clean_data(df_O):
    # Change column type to object for column: 'Volume'
    df_O = df_O.astype({'Volume': 'int'})
    return df_O

df_O_clean = clean_data(df_O.copy())
df_O_clean.head()
```
